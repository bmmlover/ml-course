```python
def run_length_encoding(x, y=None):
    b = [x[0] + 1] + x
    e = x + [x[0] + 1]
    c = [x[i] for i, _ in enumerate(x) if b[i] != b[i + 1]]
    j = 0
    d = []
    for i, _ in enumerate(x):
        j += 1
        if e[i] != e[i + 1]:
            d.append(j)
            j = 0
    if j != 0:
        d.append(j)
    return (c, d)

def run_length_encoding_vec(x, y=None):
    diff = np.where(np.diff(x) != 0)[0]
    diff = np.append(diff, len(x) - 1)
    diff = np.insert(diff, 0, -1)
    return (x[np.insert(np.diff(x).astype(np.bool), 0, True)],
            np.diff(diff))

time1, time2 = check_time(data_size, run_length_encoding, run_length_encoding_vec, 'vector', 1)
```
